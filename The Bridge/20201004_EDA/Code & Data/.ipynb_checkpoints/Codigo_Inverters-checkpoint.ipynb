```python
#Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

sns.set_style('darkgrid')

import warnings
import datetime as dt
import matplotlib.dates as mdates

warnings.filterwarnings('ignore')
```

```python
plant_1_gd = pd.read_csv('Plant_1_Generation_Data.csv')
plant_2_gd = pd.read_csv('Plant_2_Generation_Data.csv')
```

```python
plant_1_gd['DATE_TIME'] = pd.to_datetime(plant_1_gd['DATE_TIME'], errors='coerce')
plant_1_gd['DATE'] = pd.to_datetime(plant_1_gd['DATE_TIME'].dt.date)
plant_1_gd['TIME'] = plant_1_gd['DATE_TIME'].dt.time
plant_1_gd


plant_2_gd['DATE_TIME'] = pd.to_datetime(plant_2_gd['DATE_TIME'], errors='coerce')
plant_2_gd['DATE'] = pd.to_datetime(plant_2_gd['DATE_TIME'].dt.date)
plant_2_gd['TIME'] = plant_2_gd['DATE_TIME'].dt.time
plant_2_gd
```

```python
# Inverters performance

# Plant 1
plant_1_gdinall = plant_1_gd.copy()
plant_1_gdinall['time'] = plant_1_gdinall['DATE_TIME'].dt.time
plant_1_gdinall.set_index('time').groupby('SOURCE_KEY')['DC_POWER'].plot(style='o',legend=True,figsize=(20,10))
plt.title('DC POWER BY INVERTER Plant 1',size=17)
plt.ylabel('DC POWER ( kW )',color='navy',fontsize=17)
plt.show()

# Plant 2
plant_2_gdinall = plant_2_gd.copy()
plant_2_gdinall['time'] = plant_2_gdinall['DATE_TIME'].dt.time
plant_2_gdinall.set_index('time').groupby('SOURCE_KEY')['DC_POWER'].plot(style='o',legend=True,figsize=(20,10))
plt.title('DC POWER BY INVERTER Plant 2',size=17)
plt.ylabel('DC POWER ( kW )',color='navy',fontsize=17)
plt.show()
```

```python
plant_1_gdi=plant_1_gd.copy()
plant_1_gdi['time']=plant_1_gdi['DATE_TIME'].dt.time
plant_1_gdi=plant_1_gdi.groupby(['time','SOURCE_KEY'])['DC_POWER'].mean().unstack()

plant_2_gdi=plant_2_gd.copy()
plant_2_gdi['time']=plant_2_gdi['DATE_TIME'].dt.time
plant_2_gdi=plant_2_gdi.groupby(['time','SOURCE_KEY'])['DC_POWER'].mean().unstack()

cmap = sns.color_palette("Spectral", n_colors=12)

fig,ax=plt.subplots(ncols=2,nrows=1,dpi=100,figsize=(22,8))
plant_1_gdi.iloc[:,0:22].plot(ax=ax[0],color=cmap)
plant_2_gdi.iloc[:,0:22].plot(ax=ax[1],color=cmap)

ax[0].set_title('Inverters Performance Plant 1')
ax[0].set_ylabel('DC POWER ( kW )',fontsize=17,color='navy')
ax[1].set_title('Inverters Performance Plant 2')
plt.show()
```
