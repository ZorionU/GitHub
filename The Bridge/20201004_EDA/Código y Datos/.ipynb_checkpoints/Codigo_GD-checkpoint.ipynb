```python
#DATE_TIME: Date and time for each observation. Observations recorded at 15 minute intervals.
#PLANT_ID: Plant ID - this will be common for the entire file. 4135001.
#SOURCE_KEY: Source key in this file stands for the inverter id.
#DC_POWER: Amount of DC power generated by the inverter (source_key) in this 15 minute interval. Units - kW.
#AC_POWER: Amount of AC power generated by the inverter (source_key) in this 15 minute interval. Units - kW.
#DAILY_YIELD: Daily yield is a cumulative sum of power generated on that day, till that point in time.
#TOTAL_YIELD: This is the total yield for the inverter till that point in time.
```

```python
# Compute 22 inverters by date_time
plant_1_gd = plant_1_gd.groupby('DATE_TIME')[['DC_POWER','AC_POWER', 'DAILY_YIELD','TOTAL_YIELD']].agg(sum).reset_index()
plant_2_gd = plant_2_gd.groupby('DATE_TIME')[['DC_POWER','AC_POWER', 'DAILY_YIELD','TOTAL_YIELD']].agg(sum).reset_index()
```

```python
# DC Power generation on each plant

# Plant 1
plant_1_gd.plot(x= 'TIME', y='DC_POWER', style='.', figsize = (15, 8))
plant_1_gd.groupby('TIME')['DC_POWER'].agg('mean').plot(legend=True, colormap='Reds_r')
plt.ylabel('DC Power (kW)')
plt.title('DC POWER Plant 1')
plt.show()

# Plant 2
plant_2_gd.plot(x= 'TIME', y='DC_POWER', style='.', figsize = (15, 8))
plant_2_gd.groupby('TIME')['DC_POWER'].agg('mean').plot(legend=True, colormap='Reds_r')
plt.ylabel('DC Power (kW)')
plt.title('DC POWER Plant 2')
plt.show()
```

```python
# AC Power generation on each plant

# Plant 1
plant_1_gd.plot(x= 'TIME', y='AC_POWER', style='.', figsize = (15, 8))
plant_1_gd.groupby('TIME')['AC_POWER'].agg('mean').plot(legend=True, colormap='Reds_r')
plt.ylabel('AC Power (kW)')
plt.title('AC POWER Plant 1')
plt.show()

# Plant 2
plant_2_gd.plot(x= 'TIME', y='AC_POWER', style='.', figsize = (15, 8))
plant_2_gd.groupby('TIME')['AC_POWER'].agg('mean').plot(legend=True, colormap='Reds_r')
plt.ylabel('AC Power (kW)')
plt.title('AC POWER Plant 2')
plt.show()
```

```python
# Conversion rate
plant_1_losses = plant_1_gd.copy()
plant_1_losses['DATE'] = plant_1_losses['DATE_TIME'].dt.date
plant_1_losses = plant_1_losses.groupby('DATE').sum()
plant_1_losses['plant_1_losses'] = plant_1_losses['AC_POWER']/plant_1_losses['DC_POWER']*100

plant_1_losses['plant_1_losses'].plot(style='o--',figsize=(17,5),label='Real Power')

plt.title('DC/AC POWER Conversion rate Plant 1',size=17)
plt.ylabel('(%)',fontsize=14,color='black')
plt.axhline(plant_1_losses['plant_1_losses'].mean(),linestyle='--',color='red',label='mean')
plt.legend()
plt.show()

# Conversion rate
plant_2_losses = plant_2_gd.copy()
plant_2_losses['DATE'] = plant_2_losses['DATE_TIME'].dt.date
plant_2_losses = plant_2_losses.groupby('DATE').sum()
plant_2_losses['plant_2_losses'] = plant_2_losses['AC_POWER']/plant_2_losses['DC_POWER']*100

plant_2_losses['plant_2_losses'].plot(style='o--',figsize=(17,5),label='Real Power')

plt.title('DC/AC POWER Conversion rate Plant 2',size=17)
plt.ylabel('(%)',fontsize=14,color='black')
plt.axhline(plant_2_losses['plant_2_losses'].mean(),linestyle='--',color='red',label='mean')
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots(ncols=2,nrows=1,dpi=100,figsize=(20,5))

# daily yield plot
plant_1_gd.plot(x='DATE_TIME',y='DAILY_YIELD',color='navy',ax=ax[0])

# AC & DC power plot
plant_1_gd.set_index('TIME').drop('DATE_TIME',1)[['AC_POWER','DC_POWER']].plot(style='o',ax=ax[1])

ax[0].set_title('DAILY YIELD Plant 1',)
ax[1].set_title('AC/DC POWER correlation Plant 1')
ax[0].set_ylabel('(kW)',color='navy',fontsize=17)
plt.show()
```

```python
fig,ax = plt.subplots(ncols=2,nrows=1,dpi=100,figsize=(20,5))

# daily yield plot
plant_2_gd.plot(x='DATE_TIME',y='DAILY_YIELD',color='navy',ax=ax[0])

# AC & DC power plot
plant_2_gd.set_index('TIME').drop('DATE_TIME',1)[['AC_POWER','DC_POWER']].plot(style='o',ax=ax[1])

ax[0].set_title('DAILY YIELD Plant 2',)
ax[1].set_title('AC/DC POWER correlation Plant 2')
ax[0].set_ylabel('kW',color='navy',fontsize=17)
plt.show()
```

```python
# We compute a daily yield for each date for Plant 1.
plant_1_gd_daily = plant_1_gd.groupby('DATE')['DAILY_YIELD'].agg('sum')
plant_1_gd_daily.plot.bar(figsize=(15,5), legend=True)
plt.title('DAILY YIELD Plant 1')
plt.show()

# We compute daily yield for each date for Plant 2
plant_2_gd_daily = plant_2_gd.groupby('DATE')['DAILY_YIELD'].agg('sum')
plant_2_gd_daily.plot.bar(figsize=(15,5), legend=True)
plt.title('DAILY YIELD Plant 2')
plt.show()
```

```python
# We compute a total yield for each date for Plant 1.
plant_1_gd_total = plant_1_gd.groupby('DATE')['TOTAL_YIELD'].agg('sum')
plant_1_gd_total.plot.bar(figsize=(15,5), legend=True)
plt.title('TOTAL YIELD Plant 1')
plt.show()

# We compute total yield for each date for Plant 2
plant_2_gd_total = plant_2_gd.groupby('DATE')['TOTAL_YIELD'].agg('sum')
plant_2_gd_total.plot.bar(figsize=(15,5), legend=True)
plt.title('TOTAL YIELD Plant 2')
plt.show()
```

```python
#Okay, we are going to see dc power in each day produced by Plant.
#we create calendar_dc data how in each day Plant produce a dc power in each time.

calendar_dc = plant_1_gd.pivot_table(values='DC_POWER', index='TIME', columns='DATE')

# define function to multi plot

def multi_plot(data= None, row = None, col = None, title='DC Power'):
    cols = data.columns # take all column
    gp = plt.figure(figsize=(20,20)) 
    
    gp.subplots_adjust(wspace=0.2, hspace=0.8)
    for i in range(1, len(cols)+1):
        ax = gp.add_subplot(row,col, i)
        data[cols[i-1]].plot(ax=ax, style = 'k.')
        ax.set_title('{} {}'.format(title, cols[i-1]))
        
multi_plot(data=calendar_dc, row=9, col=4)
```

```python
#Okay, we are going to see dc power in each day produced by Plant.
#we create calendar_dc data how in each day Plant produce a dc power in each time.

calendar_dc = plant_2_gd.pivot_table(values='DC_POWER', index='TIME', columns='DATE')

# define function to multi plot

def multi_plot(data= None, row = None, col = None, title='DC Power'):
    cols = data.columns # take all column
    gp = plt.figure(figsize=(20,20)) 
    
    gp.subplots_adjust(wspace=0.2, hspace=0.8)
    for i in range(1, len(cols)+1):
        ax = gp.add_subplot(row,col, i)
        data[cols[i-1]].plot(ax=ax, style = 'k.')
        ax.set_title('{} {}'.format(title, cols[i-1]))
        
multi_plot(data=calendar_dc, row=9, col=4)
```

```python
# Relation btw DC Power and Daily Yield 

ax =plant_1_gd.groupby(['TIME','DATE'])['DC_POWER'].mean().unstack().plot(sharex=True,subplots=True,layout=(17,2),figsize=(20,30))
plant_1_gd.groupby(['TIME','DATE'])['DAILY_YIELD'].mean().unstack().plot(sharex=True,subplots=True,layout=(17,2),figsize=(20,20),style='-.',ax=ax)

i=0
for a in range(len(ax)):
    for b in range(len(ax[a])):
        cols = plant_1_gd.columns # take all column
        ax[a,b].set_title(cols.columns[i],size=15)
        ax[a,b].legend(['DC_POWER','DAILY_YIELD'])
        i=i+1

plt.tight_layout()
plt.show()
```
